```python
import random

while True:
    try:
        print('_' * 120)
        pass_ = '!#%246'
        passw = input('\nEnter password or access will be denied!: ')

        if pass_ == passw:
            print('\nAccess Granted!')
            name = input('\nEnter your name: ')
            gender = input('\nEnter your gender (male/female/rather not say): ')

            if gender == 'male':
                x = 'Mr.'
            elif gender == 'female':
                x = 'Mrs.'
            elif gender == 'rather not say':
                x = 'None'
            else:
                print('Enter correct gender (male/female/rather not say)')
                continue  # Skip to the next iteration if gender is invalid

            alp = 'abcdefghijklmnopqrstuvwxyz'
            num = '1234567890'
            charc = '!#$%/\\|&*'
            choice1 = random.choice(alp)
            choice2 = random.choice(num)
            choice3 = random.choice(charc)
            leng = int(input('\nWhat is your required length for your new password: '))

            # Generate the password with the desired length
            password_chars = [random.choice(alp) for _ in range(leng - 3)]
            password_chars = [choice1, choice2, choice3] + password_chars
            random.shuffle(password_chars)
            password = ''.join(password_chars)

            print(f'\n{x}{name}!\nYour new password is: {password}@gmail.com')

        else:
            print('\nIncorrect Password!\nAccess Denied.')

    except ValueError:
        print('Enter a valid length! It must be an integer.')

    run = input('\nDo you want to run again (y/n)?: ')

    if run.lower() != 'y':
        print('You chose to exit!')
        break
```
